```python
wine_data = data.iloc[:,1:]
wine_data
```

```python
wine_data.columns
```

```python
from sklearn.preprocessing import MinMaxScaler

scaled = MinMaxScaler()
wine_norm = scaled.fit_transform(wine_data)

wine_norm = pd.DataFrame(wine_norm, columns=[['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue',
       'Dilution', 'Proline']])
wine_norm
```

```python
from sklearn.preprocessing import StandardScaler

scaled = StandardScaler()
wine_scaled = scaled.fit_transform(wine_data)

wine_scaled = pd.DataFrame(wine_norm, columns=[['Alcohol', 'Malic', 'Ash', 'Alcalinity', 'Magnesium', 'Phenols',
       'Flavanoids', 'Nonflavanoids', 'Proanthocyanins', 'Color', 'Hue',
       'Dilution', 'Proline']])
wine_scaled
```

# K Means

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()
visualizer = KElbowVisualizer(model, k=(2,10), timing=True)
visualizer.fit(wine_scaled)
visualizer.show()
```

### Taking k=3 and building a custering model

```python
clus3 = KMeans(n_clusters=3, max_iter=10, random_state=1)
clus_labels = clus3.fit_predict(wine_scaled)

kdata = data
kdata['Custers'] = clus_labels
kdata.head()
```

```python
data
```

```python
data.Custers.value_counts()
```

# Hierarchial clustering

```python
import scipy.cluster.hierarchy as sch
from sklearn.cluster import AgglomerativeClustering
from sklearn.preprocessing import normalize
```

```python
plt.figure(figsize=(15,7))
plt.title('Dendogram')
dendrogram = sch.dendrogram(sch.linkage(wine_norm, 'complete'))
```

```python
plt.figure(figsize=(15,7))
plt.title('Dendogram')
dendrogram = sch.dendrogram(sch.linkage(wine_scaled, 'complete'))
```

```python
hclusters=AgglomerativeClustering(n_clusters=3,affinity='euclidean',linkage='ward')
hclusters
```

```python
y=pd.DataFrame(hclusters.fit_predict(wine_norm),columns=['clustersid'])
y['clustersid'].value_counts()
```

# PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=3)
pca_values = pca.fit_transform(wine_scaled)
pca_values.shape
```

```python
pca.explained_variance_ratio_  #top 3 relevant features
```

```python
PCA_data = pd.DataFrame(pca_values, columns=['PCA1', 'PCA2', 'PCA3'])
PCA_data
```
