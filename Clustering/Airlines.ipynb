```python
data_norm = data.iloc[:, 1:]

from sklearn.preprocessing import MinMaxScaler

model = MinMaxScaler()
data_norm = model.fit_transform(data_norm)
```

```python
data_norm = pd.DataFrame(data_norm)
```

```python
data_norm.rename(columns={0:'Balance', 1:'Qual_miles', 2:'cc1_miles', 3:'cc2_miles', 4:'cc3_miles',
       5:'Bonus_miles', 6:'Bonus_trans', 7:'Flight_miles_12mo', 8:'Flight_trans_12',
       9:'Days_since_enroll', 10:'Award?'}, inplace=True)
data_norm.head()
```

```python
data_norm.describe()
```

```python
import scipy.cluster.hierarchy as shc
from matplotlib import pyplot
pyplot.figure(figsize=(18, 7))  
pyplot.title("Dendrograms")  
dend = shc.dendrogram(shc.linkage(data_norm, method='ward'))
```

```python
from sklearn.cluster import KMeans
from yellowbrick.cluster import KElbowVisualizer

model = KMeans()

visualizer = KElbowVisualizer(model, k=(2,11), timings= True)
visualizer.fit(data_norm)        # Fit data to visualizer
visualizer.show()
```

```python
model = KMeans(n_clusters = 5, max_iter=5, random_state=1)
data_labels = model.fit_predict(data_norm)

model.labels_
```

```python
data['Cluster_lables'] = model.labels_
data.head()
```

```python
sns.lmplot( 'Balance','Bonus_miles',  data=data,hue = 'Cluster_lables', fit_reg=False, size = 6 )
plt.show()
```
